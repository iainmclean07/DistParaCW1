```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

go_medians_DS1 = [16.162, 8, 4.18, 2.186, 1.12, 0.743, 0.750, 0.736]
c_medians_DS1 = [15.288, 7.661, 4.092, 2.097, 1.073, 0.688, 0.678, 0.674]

go_medians_DS2 = [68.391, 34.313, 17.813, 9.232, 4.651, 3.093, 3.083, 3.078]
c_medians_DS2 = [65.638, 33.104, 17.163, 8.937, 4.514, 2.829, 2.812, 2.808]

go_medians_DS3 = [39.251, 19.675, 12.916, 12.909, 12.909]
c_medians_DS3 = [38.054, 19.118, 11.897, 11.850, 11.854]
```

```python
figure(figsize=(10, 10))

N = 8

ind = np.arange(N) 
width = 0.35       
plt.plot(ind, go_medians_DS1, label='Go')
plt.plot(ind, c_medians_DS1, label='C')

plt.ylabel('Runtime (s)')
plt.xlabel('Threads')
plt.title('DS1 Runtimes')

plt.xticks(ind + width / 2, ('1', '2', '4', '8', '16', '32', '48', '64'))
plt.legend(loc='best')
plt.show()
```

```python
figure(figsize=(10, 10))

N = 8

ind = np.arange(N) 
width = 0.35       
plt.plot(ind, go_medians_DS2, label='Go')
plt.plot(ind, c_medians_DS2, label='C')

plt.ylabel('Runtime (s)')
plt.xlabel('Threads')
plt.title('DS2 Runtimes')

plt.xticks(ind + width / 2, ('1', '2', '4', '8', '16', '32', '48', '64'))
plt.legend(loc='best')
plt.show()
```

```python
figure(figsize=(10, 10))

N = 5

ind = np.arange(N) 
width = 0.35       
plt.plot(ind, go_medians_DS3, label='Go')
plt.plot(ind, c_medians_DS3, label='C')

plt.ylabel('Runtime (s)')
plt.xlabel('Threads')
plt.title('DS3 Runtimes')

plt.xticks(ind + width / 2, ('8', '16', '32', '48', '64'))
plt.legend(loc='best')
plt.show()
```

```python
figure(figsize=(10, 10))

c_t1s = 15.795
go_t1s = 15.963

labels = [1, 2, 4, 8, 16, 32, 48, 64]

c_absolute_speedups = [c_t1s/x for x in c_medians_DS1]
go_absolute_speedups = [go_t1s/x for x in go_medians_DS1]

print("c", c_absolute_speedups)
print("go", go_absolute_speedups)


N = 8

ind = np.arange(N) 
width = 0.27       
plt.bar(ind, c_absolute_speedups, width, label='C')
plt.bar(ind + width, go_absolute_speedups, width, label='Go')
plt.bar(ind + width*2, labels, width, label='Ideal')

plt.ylabel('Speedup')
plt.xlabel('Threads')
plt.title('DS1 Absolute Speedup')

plt.xticks(ind + width / 2, labels)
plt.legend(loc='best')
plt.show()
```
